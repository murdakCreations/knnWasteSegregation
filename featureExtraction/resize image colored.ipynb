```python
strArr = ""
arr = []
def combRGB(indx1, indx2):
    global strArr
    #print(indx1, indx2)
    for indx3 in range(0, resizedImg.shape[2]):
        strArr += str(resizedImg[indx1][indx2][indx3])
    if stackRGB(strArr, arr) == False:
        arr.append(strArr)
        
    strArr = ""

def stackRGB(strArr, arr):
    for indx in range(len(arr)):
        if arr[indx] == strArr:
            return True
    return False

for indx1 in range(0, resizedImg.shape[0]):
    for indx2 in range(0, resizedImg.shape[1]):
        combRGB(indx1, indx2)
print(arr)
print(len(arr))
```
