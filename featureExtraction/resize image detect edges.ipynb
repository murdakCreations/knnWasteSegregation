```python
from skimage.feature import canny
from skimage.io import imread, imshow
import matplotlib.pyplot as plt

# double dash to avoid syntax error
image = imread("C:/Users/Blobby/Desktop/glass/glass120.jpg", as_gray=True)
imshow(image)
plt.show()
```

```python
import cv2
resizedImg = cv2.resize(image,(30,30))
plt.imshow(resizedImg, cmap = 'gray')
plt.show()
```

```python
resizedImgEdges = canny(resizedImg)
imshow(resizedImgEdges)
plt.show()
```
